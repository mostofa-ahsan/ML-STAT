```python
import numpy as np
import pandas as pd
import scipy
import math
# from scipy.stats import chisqprob
# from scipy import chi2
from scipy import stats
import math
from math import sqrt
from itertools import product
import itertools
from itertools import combinations
from scipy.stats import chi2


df= pd.read_excel("C:/Users/DOTSC75/Desktop/Car_data.xlsx")
df1= df[['buying', 'mainta', 'doors', 'persons', 'lug_bot', 'safety']]

df1.head(5)

# columns= df.columns
# df.columns
# df.at[2,'buying']
ADC=dict()
N_Gain=dict()
S_U=dict()
M_I= dict()

for var1 in df1.columns:
    p = df[var1].unique()
#     print(type(var1))
#     x_cols = [x for x in df.columns if x != var1]
#     print (var1)
#     for var2 in x_cols:
    q = df['acceptability'].unique()
#         print(var2)
    result=[]
    for i in range(len(p)):
        row=[]
        
        for j in range(len(q)):
            count=0
                
            for d in range(0, len(df)):
                if((df.at[d,var1]==p[i]) & (df.at[d,'acceptability']==q[j])):
                    count=count+1

            row.append(count)
   
        result.append(row)

    np.array(result).reshape(len(p),len(q) )
    
    total= sum(map(sum, result))
 
    p_matrix=[]
    for i in range (len(p)):
        p_row=[]
        for j in range(len(q)):
            prb= (result[i][j])/total
            p_row.append(prb)
        p_matrix.append(p_row)
    p1_mat=np.array(p_matrix).reshape(len(p),len(q) )
  
    H_f=0
    for i in range (len(p)):
        e=(sum(p1_mat[i,:]))
        if e!=0:            
            H_f= H_f+ (e * math.log2(e**(-1)))
    
    H_c=0
    for i in range (len(q)):
        g=sum(p1_mat[:,i])
        if g!=0:            
            H_c= H_c+ (g * math.log2(g**(-1)))


    H_cf=0
    for i in range (len(p)):
        for j in range(len(q)):
            k= (p1_mat[i][j])
            if k!=0:
                H_cf= H_cf+ (k * math.log2(k**(-1)))

    I_cf= H_f + H_c -H_cf   
    adc= I_cf/H_c
    ADC[var1]=adc    
    n_gain=I_cf/H_f
    N_Gain[var1]=n_gain    
    s_u=2*I_cf/(H_c+H_f)
    S_U[var1]=s_u    
    M_I[var1]=I_cf    
ADC_s = {k: v for k, v in sorted(ADC.items(), key=lambda x: x[1], reverse=True)}    
N_Gain_s = {k: v for k, v in sorted(N_Gain.items(), key=lambda x: x[1], reverse=True)} 
S_U_s = {k: v for k, v in sorted(S_U.items(), key=lambda x: x[1], reverse=True)} 
M_I_s = {k: v for k, v in sorted(M_I.items(), key=lambda x: x[1], reverse=True)}  

print('Asymetric Dependency Coefficient RANK:')                
print(ADC_s)
print('Normalized Gain RANK:')  
print(N_Gain_s)
print('Symmetrical Uncertainty RANK:')  
print(S_U_s)
print('Mutual Information RANK:')  
print(M_I_s)

```
