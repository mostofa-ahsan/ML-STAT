```python
import numpy as np
import math
import operator
import pandas as pd
from operator import itemgetter
from collections import OrderedDict


Xtrain= np.loadtxt("irisTraining.txt")
Xtest = np.loadtxt("irisTesting.txt")

tr=len(Xtrain)
ts=len(Xtest)
n = Xtrain.shape[0]
d = Xtrain.shape[1]-1

nt=Xtest.shape[0]
dt=Xtest.shape[1]-1
#print(tr)
#print(ts)
train=Xtrain[:,0:d]
test = Xtest[:,0:dt]

train_label=np.array(Xtrain[:,4])
test_label=np.array(Xtest[:,4])
#print(train_label[2])
TrainRow=np.array(train)
TestRow = np.array(test)

#print(train)
TestLabel=[]
def euclideanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)
#from scipy import scipy.spatial.distance.euclidean
for i in range(0,ts):
    a= (TestRow[i,:])
    distance=[]
    position = []
    for j in range(0,tr):
        
        
        b= (TrainRow[j,:])
        dst = euclideanDistance(a, b,2)
        #print(dst)
        k=5
        distance.append(dst)
        position.append(j)
    
        
        
    dict1= dict(zip(position,distance))
    #sorted(dict1.iteritems())
    d = OrderedDict(sorted(dict1.items(), key=itemgetter(1)))
    #sorted(dict1.values())
    d1=d.keys()
    top_k=[]
    for m in d1:
        top_k.append(m)
    top_k=top_k[:5]
    #print(top_k)
    
    trainclass=[]
    for c in top_k:
        #print(c)
        #print(train_label[c])
        #print()
        trainclass.append(train_label[c])
        
    #print(trainclass)
    def most_frequent(List): 
        return max(set(List), key = List.count) 
  
     
    p=(most_frequent(trainclass))
    TestLabel.append(p)
#print(TestLabel[2])
#print(test_label[2])
acc=0
for l in range(0,len(test_label)):
    if test_label[l]==TestLabel[l]:
        acc=acc+1
print((acc/len(TestLabel))*100)
print(len(TestLabel))

    


    
    
```
