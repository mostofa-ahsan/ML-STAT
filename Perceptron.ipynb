```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import math
import operator


df = pd.read_excel("D:/Dataset/Perceptron.xlsx",header=None)
df.head()
```

```python
w_0=0
w_1=0
w_2=0
l_r=.001
w_j=np.array((w_0,w_1,w_2))
# w_j=np.c_[w_j]
w_j=w_j.transpose()
cf_diag_sum=len(df)-1
epoch=0
while (cf_diag_sum!=len(df)):
    plus_one=0
    minus_one=0
    w_j=w_j
    print(w_j)
    for i in range(len(df)):
        x_j=np.array((i+1,df.at[i,0],df.at[i,1]))
       
        a= x_j.dot(w_j)
        if a>=0:
            y_h=1
        if a<0:
            y_h=-1

        y=df.at[i,2]
        if (y==1) and (y_h==1):
            plus_one=plus_one+1
        if (y==-1) and (y_h==-1):
            minus_one=minus_one+1
        d_w_j=(x_j.dot(l_r*(y-y_h)))
        w_j=w_j+d_w_j
        
    print(w_j)
    cf_diag_sum=plus_one+minus_one
    epoch=epoch+1
    print(cf_diag_sum, epoch)
```
